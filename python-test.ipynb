```python
# creating game class

class Game:
    __index__ = 0
    __max__ = None
    
    def __init__(self, *levels):
        self.levels = levels
        self.__max__ = len(self.levels)
        if self.__max__ <= 0:
            __index__ = None
        else:
            self.__index__ == 0
        return None

    def __iter__(self):
        return self

    def __next__(self):
        if self.__index__ == None:
            raise StopIteration
            
        if self.__index__ < self.__max__:
            value = self.levels[self.__index__]
            self.__index__ += 1
            return value
        else:
            raise StopIteration
            
    def __len__(self):
        return self.__max__
```

```python
# LEVEL-WISE Breakdown:
# ✅ LEVEL 1 – Crack the Login
#  Aapko admin user ka password guess karna hai.
#  Valid passwords passwords.txt file mein hote hain.
#  Aapko 3 chances milti hain.
#  Sahi password mil gaya → next level
#  Galat → game over

def LevelOne(): 
    # create a passwords.txt file with valid password
    validPassword = "3"
    try:
        with open("./passwords.txt", "x") as passwordsFile:
            passwordsFile.write(validPassword)
    except FileExistsError:
        with open("./passwords.txt", "r") as passwordsFile:
            validPassword = passwordsFile.read()

    for i in range(3):
       guessPassword = input("Enter Password: ")
       if guessPassword == validPassword:
           print("Congratulations You Win This Level")
           return True
       else:
           print("Failed, Try Again")
            
        
   
    return False
    
```

```python
# LEVEL 2 – Trace the Hidden Keyword
#  Aapko ek file clue.txt se keyword accesskey find karna hota hai.
#  User se direct poocha jata hai keyword kya hai.
#  Agar:

#  Input correct hai
#  AND 30 seconds ke andar diya gaya hai
# → next level
# warna → fail

import random
from datetime import datetime
import time


def LevelTwo():
    fruit_names ="Apple\nBanana"
    try:
        with open("./clue.txt", "x") as clue:
            clue.write(fruit_names)
    except FileExistsError:
        with open("./clue.txt", "r") as clue:
            fruit_names = clue.read()
    fruit_list = fruit_names.split("\n")
    
    print("Guess Fruit Name")
    print("******************") 
    random_fruit = random.choice(fruit_list)
    
    txt = f'A Fruit Name startsWith {random_fruit[0]} and ends with {random_fruit[-1]}, Guess Name In 30sec \n--> '
    t1 = datetime.now()
    user_input = input(txt)
    t2 = datetime.now()

    if int((t2 - t1).total_seconds()) > 30: 
        print("Timeout")
        return False
    elif user_input.strip() == random_fruit:
        print("Congratulations You Win This Level")
        return True
    else:
        print("Invalid Name")
        return False
        
```

```python
# LEVEL 3 – Decode the Cipher
#  File encrypted.txt mein ek encrypted message hota hai (e.g., Caesar cipher).
#  Aapko ise decrypt kar ke user se poocha jata hai correct message kya hai.
#  Agar sahi diya → next level

from caesarcipher import CaesarCipher


def LevelThree():
    cipher = CaesarCipher("encryption", offset=3)
    defaultCipher = cipher.encoded
    try:
        with open("./encrypted.txt", "x") as cipher:
            cipher.write(defaultCipher)
    except FileExistsError:
        with open("./encrypted.txt", "r") as cipher:
            defaultCipher = cipher.read()

    message = input(f'Tell Us Correct Message\n {defaultCipher}\n->')
    if(CaesarCipher(message, offset=3).encoded == defaultCipher):
        print("Congratulations You Win This Level")
        return True
    else:
        print("Invalid Message")
        return False
```

```python
# LEVEL 4 – Defuse the Bomb (Logic Puzzle)
#  Ek riddle diya jata hai:
#  “I speak without a mouth and hear without ears. What am I?”
#  Correct answer: echo
#  3 chances hoti hain.
#  Agar sahi answer → bomb.txt mein likha jata hai: “Bomb defused successfully!”
#  Galat → “Bomb exploded!”


def LevelFour():
    riddle = "I speak without a mouth and hear without ears. What am I?"
    correct_answer = "echo"
    for attempt in range(3):
        user_answer = input(f"{riddle}\nAttempt {attempt+1}/3 -> ").strip().lower()
        if user_answer == correct_answer:
            with open("./bomb.txt", "w") as f:
                f.write("Bomb defused successfully!")
            print("Congratulations You Win This Level")
            return True
        else:
            print("Wrong answer. Try again.")
    with open("./bomb.txt", "w") as f:
        f.write("Bomb exploded!")
    print("Bomb exploded!")
    return False
```

```python
# LEVEL 5 – Final Firewall
#  Grid (matrix) mein ek weak point hidden hota hai e.g. (2,2)
#  User se grid coordinates liye jate hain.
#  Agar coordinates match karein → firewall breached → mission complete
# Concepts used:
# 2D list, tuple matching, input, try-except
def LevelFive():
    weak_point = (2, 2)
    try:
        raw = input("Enter coordinates as row,col (1-indexed): ")
        parts = raw.split(",")
        row = int(parts[0].strip())
        col = int(parts[1].strip())
    except Exception:
        print("Invalid format. Use row,col like 2,2")
        return False

    if (row, col) == weak_point:
        print("Firewall breached – mission complete!")
        return True
    else:
        print("Access denied – wrong coordinates.")
        return False
```

```python
game = Game(LevelOne, LevelTwo, LevelThree, LevelFour, LevelFive)
print("Welcome to Digital Game Arena")
print(f'\nThere are {len(game)} Level to Win This Game\n')

for each_game in game:
    gameStatus = each_game()
    if gameStatus == True:
        print("\n\n-------------------------")
        print("Entering to Next Level")
        print("------------------------\n\n")
    else:
        print("\n\nGame Over, Better Luck Next Time :)")
        break


print("\n\nGreat You Have Win All Level")
```
