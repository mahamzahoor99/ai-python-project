```python
import pandas as pd

# Load the Titanic dataset and display the first 10 rows
titanic=pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
print(titanic.head(10))
```

```python
# Check the number of rows and columns in the dataset

print(titanic.shape)
print("No. of rows: ",len(titanic),"\t|\tNo. of colums: ",len(titanic.columns))
```

```python
# Display the column names and their data types

# import numpy as np
# print("Names of colums: ",np.array(titanic.columns))
list_titanic=list(titanic.columns)
print("Names of colums: ",list_titanic)
dtypes=pd.DataFrame(titanic.dtypes)
dtypes=dtypes.replace({'int64':'integer','object':'string','float64':'floating number'})
print(dtypes)
```

```python
# Show the total number of missing values in each column

# titanic.isna().sum()
NaN_count=len(titanic)-titanic.count()
print(NaN_count)
```

```python
# Fill missing values in the Age column with the average age

import numpy as np
# avg_age=titanic.Age.mean()
avg_age=titanic['Age'].sum()/titanic['Age'].count() # similar to: titanic['Age'].sum()/titanic.Age.count()
print("average age: ", avg_age)
print("Age column with replaced values of average ages:\n",titanic.Age.replace(np.nan,avg_age))
```

```python
# Count how many passengers survived and how many did not

print("passengers survived: ",titanic.Survived[titanic.Survived == 1].count())
print("passengers died: ",titanic.Survived[titanic.Survived == 0].count())
```

```python
# Find the survival rate (percentage) of male and female passengers separately

surv_pas=titanic[titanic.Survived==1][['Survived','Sex']]
surv_male=surv_pas[surv_pas.Sex=='male'].Sex.count()
surv_female=surv_pas[surv_pas.Sex=='female'].Sex.count()
print("total survived: ",surv_pas.Survived.count())
print("male: ",surv_male)
print("female: ",surv_female)
print("Male survival rate: ",((surv_male/surv_pas.Survived.count())*100).round(2))
print("Female survival rate: ",((surv_female/surv_pas.Survived.count())*100).round(2))
```

```python
# Display the average age of passengers who survived vs. who did not survive

titanic[["Survived","Age"]]
surv_avg_age=((titanic.Age[titanic.Survived==1]).sum())/((titanic.Age[titanic.Survived==1]).count())
print("averag age of passengers survived: ",surv_avg_age.round(2))
died_avg_age=((titanic.Age[titanic.Survived==0]).sum())/((titanic.Age[titanic.Survived==0]).count())
print("averag age of passengers diec: ",died_avg_age.round(2))
```

```python
# Find the most common embarkation port (Embarked column)

# print(titanic.Embarked.unique())

print("Most common embarkation port: ",titanic["Embarked"].mode())
```

```python
# Show the total number of passengers in each passenger class (Pclass)

print("Pclass entries: ",titanic.Pclass.unique())

print("Pclass 3: ",titanic.Pclass[titanic.Pclass==3].count())
print("Pclass 1: ",titanic.Pclass[titanic.Pclass==1].count())
print("Pclass 2: ",titanic.Pclass[titanic.Pclass==2].count())
```

```python
# Display all passengers whose fare is greater than 100

titanic[titanic.Fare>100]
```

```python
# Find the highest fare and the passenger(s) who paid it

# titanic.Fare.max()

titanic.nlargest(5,"Fare")
```

```python
# Group the data by Pclass and find the average fare of each class

print('List of total passenger classes: ',np.sort(titanic.Pclass.unique()))
pclass1=titanic[(titanic.Pclass==1)]
pclass2=titanic[titanic.Pclass==2]
pclass3=titanic[titanic.Pclass==3]
print('Average fare of Pclass_1 is ',pclass1.Fare.mean(numeric_only=True).round(2))
print('Average fare of Pclass_2 is ',pclass2.Fare.mean(numeric_only=True).round(2))
print('Average fare of Pclass_3 is ',pclass3.Fare.mean(numeric_only=True).round(2))
```

```python
# Create a new column FamilySize = SibSp + Parch + 1 and display top 10 rows

titanic['FamilySize']=(titanic.SibSp.dropna()+titanic.Parch.dropna()+1)
titanic.head(10)
```

```python
# Sort the dataset by age in descending order and show the top 5 oldest passengers

titanic.sort_values(by='Age',ascending=False).head(5)
```
